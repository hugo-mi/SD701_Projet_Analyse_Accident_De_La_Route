#### On applique l'alorithme Apriori

```python
%%time
#!pip install apyori
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

records = []
for i in range(200):
    records.append([str(df_var_select_accident_grave.values[i,j]) for j in range(0, 13)])
    
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=4, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
output = list()
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    
    value0 = str(items[0])
    value1 = str(items[1])
    
    value2 = str(item[1])[:7]
    
    value3 = str(item[2][0][2])[:7]
    value4 = str(item[2][0][3])[:7]
    
    rows = (value0, value1, value2, value3, value4)
    output.append(rows)
    
labels = ["Antecedents", "Consequences", 'Support', 'Confidence', 'Lift']
accident_rules = pd.DataFrame.from_records(output, columns = labels)
accident_rules
```

```python
accident_rules_bis = accident_rules.loc[accident_rules["Consequences"] == "BlÃ©ssÃ© grave et/ou mort"]

accident_rules_bis["Support"] = accident_rules_bis["Support"].astype(float)
accident_rules_bis["Confidence"] = accident_rules_bis["Confidence"].astype(float)
accident_rules_bis["Lift"] = accident_rules_bis["Lift"].astype(float)
```

```python
accident_rules_sorted = accident_rules_bis.sort_values('Lift', ascending=False)
accident_rules_sorted = accident_rules_sorted.drop_duplicates()
accident_rules_sorted.head(50)
```
